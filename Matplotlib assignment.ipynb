### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
#A bar plot, also known as a bar chart, is a type of chart that represents categorical data with rectangular bars. The length or height of each bar is proportional to the value it represents. Bar plots are commonly used to compare and display the distribution or relationship between different categories of data.
```

```python
#Bar plots are used for the following purposes:

#Comparison: Bar plots are effective for comparing the values of different categories, making it easy to identify the largest or smallest values among the categories.

#Categorical Data Visualization: Bar plots are suitable for visualizing categorical data, such as the sales of different products, the performance of teams in a competition, or the distribution of votes in an election.

#Frequency Distribution: Bar plots can display the frequency of occurrence of different categories in a dataset.

#Time-Series Data: When dealing with time-series data, bar plots can represent data for each time point, allowing for easy trend analysis.
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
#create a bar plot
plt.figure(figsize=(8,6))
plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits')
plt.show()
# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit)  # Use barh() for horizontal bars
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Company Profits')
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
#A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the five-number summary of a dataset, which consists of the minimum value, first quartile (Q1), median (Q2), third quartile (Q3), and maximum value. The box plot is called so because the plot consists of a box and whiskers.
```

```python
#Box plots are used for several reasons:

#Visualizing data distribution: Box plots provide a concise and clear way to understand the spread, skewness, and central tendency of the data. They offer a quick summary of key statistical measures.

#Identifying outliers: Box plots help in identifying potential outliers in the dataset, which could be data points that are far from the majority of the data.

#Comparing distributions: Box plots allow for easy comparison of multiple datasets side by side, making it simple to spot differences in their central tendencies and spreads.

#Handling large datasets: When dealing with large datasets, box plots provide a compact representation of the data's distribution, making it easier to grasp the overall picture without examining each individual data point.

#Handling non-normal distributions: Box plots are robust to non-normal distributions and are less affected by extreme values than other graphical representations like histograms.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1 ,box2]

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=["Box 1", "Box 2"])
plt.show
```
